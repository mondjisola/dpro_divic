[Problem 1] Creating random numbers

```python
import numpy as np

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
random_numbers = np.random.multivariate_normal(mean, cov, size=500)
random_numbers.shape
```

---

[Problem 2] Visualization by scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

x1, x2 = random_numbers.T # <=> x1, x2 = random_numbers[:, 0], random_numbers[:, 1]
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("dummy data scatter plot")
plt.scatter(x1, x2)
plt.show()
```

---

[Problem 3] Visualization by histogram

```python
# Create figure and axes
fig, axs = plt.subplots(1, 2, figsize=(10, 5), tight_layout=True)

# x1 histogram
axs[0].set_xlabel("x1")
axs[0].set_ylabel("frequency")
axs[0].set_title("histogram of x1")
axs[0].hist(x1, bins=50)
axs[0].set_xlim(left=-5, right=2)

# x2 histogram
axs[1].set_xlabel("x2")
axs[1].set_ylabel("frequency")
axs[1].set_title("histogram of x2")
axs[1].hist(x2, bins=50)
axs[1].set_xlim(left=-5, right=2)

plt.show()
```

---

[Problem 4] Addition of data

```python
new_mean = (0, -3)
new_random_numbers = np.random.multivariate_normal(new_mean, cov, size=500)
```

```python
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("data addition scatter plot")
plt.scatter(*random_numbers.T, label='0')
plt.scatter(*new_random_numbers.T, label='1')
plt.legend()
plt.show()
```

---

[Problem 5] Data combination

```python
data = np.vstack((random_numbers, new_random_numbers))
data.shape
```

---

[Problem 6] Labeling

```python
labels = np.concatenate((np.zeros((500, 1)), np.ones((500, 1))))
data = np.hstack((data, labels))
data.shape
```
