```python
# @title Estimated

class ScratchSVMClassifier:
    """
    Scratch implementation of SVM classifier

    Parameters
    ----------
    num_iter : int
        Number of iterations
    lr : float
        Learning rate
    kernel : str
        Kernel type. Linear kernel ('linear') or polynomial kernel ('polly')
    threshold : float
        Threshold for selecting support vectors
    verbose : bool
        True to output the learning process

    Attributes
    ----------
    self.n_support_vectors : int
        Number of support vectors
    self.index_support_vectors : ndarray of shape (n_support_vectors,)
        Index of support vectors
    self.X_sv : ndarray of shape (n_support_vectors, n_features)
        Features of support vectors
    self.lam_sv : ndarray of shape (n_support_vectors, 1)
        Undetermined multiplier of support vectors
    self.y_sv : ndarray of shape (n_support_vectors, 1)
        Support vector labels
    """

    def __init__(self, num_iter, lr, kernel='linear', threshold=1e-5, verbose=False, degree=1, gamma=1, weight=0):
        # Record hyperparameters as attributes
        self.iter = num_iter
        self.lr = lr
        self.kernel = kernel
        self.threshold = threshold
        self.verbose = verbose
        self.degree = degree
        self.gamma = gamma
        self.weight = weight

    def fit(self, X, y, X_val=None, y_val=None):
        """
        Train the SVM classifier. If validation data is input, the accuracy for it is also calculated for each iteration.

        Parameters
        ----------
        X: ndarray of shape (n_samples, n_features)
            Features of training data
        y: ndarray of shape (n_samples, )
            Correct value of training data
        X_val: ndarray of shape (n_samples, n_features)
            Features of validation data
        y_val: ndarray of shape (n_samples, )
            Correct value of validation data
        """
        self.label = np.unique(y)
        y = np.where(y == self.label[0], -1, 1)
        self.lamda = np.random.rand(len(X))
        for i in range(self.iter):
            self._update_lambda(X, y)
            if self.verbose:
                print("iteration: {}, lamda: {}".format(i, self.lamda))
        self.index_support_vector = np.where(self.lamda > self.threshold)[0]
        self.X_sv = X[self.index_support_vector, :]
        self.y_sv = y[self.index_support_vector].reshape(-1, 1)
        self.lam_sv = self.lamda[self.index_support_vector].reshape(-1, 1)
        self.n_support_vectors = len(self.lam_sv)

    def predict(self, X):
        """
        Estimate labels using SVM classifier.

        Parameters
        ----------
        X: ndarray of shape (n_samples, n_features)
            Sample

        Returns
        ------
        ndarray of shape (n_samples, 1)
            SVM classifier estimation results
        """
        n = X.shape[0]
        y_pred = np.zeros(n)

        for i in range(n):
            temp = 0
            for j in range(self.n_support_vectors):
                temp += self.lam_sv[j] * self.y_sv[j] * self._kernel(X[i], self.X_sv[j])
            if temp >= 0:
                y_pred[i] = self.label[1]
            else:
                y_pred[i] = self.label[0]

        return y_pred

    def _update_lambda(self, X, y):
        """
        Update parameter lambda during training process
        Parameters
        -----------
        X: input data, ndarray (n_samples, n_features)
        y: label data, ndarray (n_samples,)
        """
        n = X.shape[0]

        for i in range(n):
            update_sum = 0
            for j in range(n):
                update_sum += self.lamda[j]*y[i]*y[j]*self._kernel(X[i],X[j])
            self.lamda[i] += self.lr*(1-update_sum)
            if self.lamda[i] < 0:
                self.lamda[i] = 0

    def _kernel(self, X_i, X_j):
        """
        Compute the result when applying kernel function to two samples

        Parameters
        ----------
        X_i, X_j: ndarray of shape (n_features,)
            Samples
        """
        if self.kernel == 'linear':
            return np.dot(X_i, X_j.T)
```

# Learning and estimation with scrarch model

```python
# @title Data loading and preprocessing

import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
y = iris.target

X = X[y != 2]
y = y[y != 2]

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

y_train = np.where(y_train == 1, -1, 1)
y_test = np.where(y_test == 1, -1, 1)

print("Forme de X_train:", X_train.shape)
print("Forme de y_train:", y_train.shape)
print("Forme de X_test:", X_test.shape)
print("Forme de y_test:", y_test.shape)
```

```python
# @title Training

svm = ScratchSVMClassifier(num_iter=1000, lr=0.01, kernel='linear', verbose=True)
svm.fit(X_train, y_train)
```

```python
# @title Estimation

y_pred = svm.predict(X_test)
y_pred
```

```python
# @title Prediction accuracy

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: {}'.format(accuracy))
```

```python
# @title Prediction precision

from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred)
print('Precision: {}'.format(precision))
```

```python
# @title Prediction recall

from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred)
print('Recall: {}'.format(recall))
```

```python
# @title Decision region vizualisation function

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def decision_region(X,y,slr):
    mesh_f0, mesh_f1  = np.meshgrid(
        np.arange(np.min(X[:,0]), np.max(X[:,0]), 0.01),
        np.arange(np.min(X[:,1]), np.max(X[:,1]), 0.01)
    )
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = slr.predict(mesh).reshape(mesh_f0.shape)
    plt.title('decision region')
    plt.xlabel('feature0')
    plt.ylabel('feature1')
    plt.contourf(mesh_f0, mesh_f1, y_pred,cmap=ListedColormap(['pink', 'skyblue']))
    plt.contour(mesh_f0, mesh_f1, y_pred,colors='red')
    plt.scatter(X[y==-1][:, 0], X[y==-1][:, 1],label='1')
    plt.scatter(X[y==1][:, 0], X[y==1][:, 1],label='2')
    plt.legend()
    plt.show()
```
