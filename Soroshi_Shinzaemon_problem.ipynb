[Problem 1] Number of grains of rice on the 100th day


```python
# Calculation of the total number of grains received on the 100th day.
# I use geometric series formula to calculate the sum of the n consecutive first terms,
# of a geometric series where the first term (a1) is 1 , the common ratio (r) is 2, and n (100) is the number of terms.
# âˆ‘ = r**n - a1
total_grains = 2 ** (100) - 1

# Output the answer.
print("The total number of grains received on the 100th day is {} grains.".format(total_grains))
```

```python
# Plot a line graph of the number of grains received per day and the total number of grains received by that day.
import matplotlib.pyplot as plt

# Calculation of the number of grains received per day.
grains_per_day = []
for i in range(1, 101):
  grains_per_day.append(2 ** (i - 1))

# Plot the graph.
plt.plot(grains_per_day, label="Grains per day")
plt.xlabel("Number of days")
plt.ylabel("Number of rice grains")
plt.title("Change in the number of rice grains received from day 1 to day 100.", fontsize=10.5)
plt.legend()
plt.show()
```

[Problem 2] Number of rice grains outside of the 100th day

```python
def compute_sorori_shinzaemon(n_days=100):
  """A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

  Parameters
  -----------
  n_days: int
      Number of days to get rice (default : 100)

  Returns
  -------
  list_n_grains : list
      List of the number of rice you get one day (list length equals n_days)
  list_total_grains : list
      List of total number of rice received by a certain day (list length is equal to n_days)
  """

  list_n_grains = []
  list_total_grains = []
  for i in range(1, n_days + 1):
    n_grains = 2 ** (i - 1)
    list_n_grains.append(n_grains)
    total_grains = sum(list_n_grains)
    list_total_grains.append(total_grains)
  return list_n_grains, list_total_grains
```

```python
# Answer how many rice grains will be received in the number of days taken we take the course.
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=32)
print("In the 32 days that we have been following the course, we will have received {} grains of rice.".format(list_total_grains[-1]))
```

```python
# Plot the graph.
plt.plot(list_n_grains, label="Grains per day")
plt.xlabel("Number of days")
plt.ylabel("Number of rice grains")
plt.title("Change in the number of rice grains received from day 1 to day 32.", fontsize=10.5)
plt.legend()
plt.show()
```

```python
# Function to calculate how many people can live for how many days with this rice.

def calculate_days_to_live(n_grains, n_people):
  """Calculates the number of days that n_people can live on n_grains of rice.

  The number of grains of rice needed per day is not provided here. Please do your own research and put it into a formula. Please leave a comment on how you did your research and what you used as your basis.

  Parameters
  -----------
  n_grains: int
      Number of grains of rice
  n_people: int
      Number of people

  Returns
  -------
  int
      Number of days that n_people can live on n_grains of rice

  Notes
  -----
  The formula used here is:
  days_to_live = n_grains / (n_people * grains_per_day_estimate)

  The estimate used here is based on the guideline of 3-4 ounces of whole grain rice per day per adult.
  """

  # Research-based estimate of grains needed per day.
  ounces_per_day_estimate_low = 3
  ounces_per_day_estimate_high = 4
  grams_per_ounce = 28.35

  # Calculate the range of grains needed per day.
  grains_per_day_low = ounces_per_day_estimate_low * grams_per_ounce
  grains_per_day_high = ounces_per_day_estimate_high * grams_per_ounce

  # Calculate the average grains per day for simplicity
  grains_per_day_estimate = (grains_per_day_low + grains_per_day_high) / 2

  # Calculate the number of days of sustenance
  days_to_live = n_grains / (n_people * grains_per_day_estimate)

  return int(days_to_live)
```
